## Using Sampling generating training and test data

```python
data=dataset.sample(frac=0.80,random_state=200)
data_unseen=dataset.drop(data.index)
data.reset_index(inplace=True,drop=True)
data_unseen.reset_index(inplace=True,drop=True)
```

```python
data=data.drop(columns=['Unnamed: 0'])
data_unseen=data_unseen.drop(columns=['Unnamed: 0'])
```

```python
from pycaret.classification import *
```

### Working with training data

```python
exp_classification=setup(data=data,target='classification',session_id=123)
```

```python
compare_models()
```

```python
et = create_model('et') # it is automatically doing cross validation
```

## Fine Tuning

```python
tuned_et = tune_model(et)
```

## Evaluate the model

```python
evaluate_model(et) #since et was better than tuned_et
```

## Predict the unseen data

```python
unseen_pred = predict_model(et,data=data_unseen)
unseen_pred.head()
```
