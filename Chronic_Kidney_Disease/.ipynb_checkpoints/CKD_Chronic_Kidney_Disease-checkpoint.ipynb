###  Better Approach

```python
# Impute all columns
imputer_mode = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
imputer_median = SimpleImputer(missing_values=np.nan, strategy='median')

# Apply imputation based on column type
for col in df.columns:
    if df[col].dtype == 'object':  # Categorical columns
        df[col] = imputer_mode.fit_transform(df[[col]])
    else:  # Numerical columns
        df[col] = imputer_median.fit_transform(df[[col]])

df
```

### Here we will classify the columns into different types so that we know where to apply label encoder(Ordinal Data) and where one hot encoding(Nominal Data) 

```python
X.columns
```

```python
def classify_features(X):
    categorical_features = []
    non_categorical_features = []
    discrete_features = []
    continuous_features = []
    
    for col in X.columns:
        if X[col].dtype == 'object':  # Check if the column is categorical
            if X[col].nunique() < 3:  # Check if the number of unique values is small
                categorical_features.append(col)
            else:
                non_categorical_features.append(col)
        elif X[col].dtype in ['int64', 'float64']:  # Check if the column is numerical
            if X[col].nunique() < 10:  # Check if the number of unique values is small
                discrete_features.append(col)
            else:
                continuous_features.append(col)
    
    return categorical_features, non_categorical_features, discrete_features, continuous_features
```

```python
categorical,non_categorical,discrete,continuous = classify_features(X)
```

## 5 Mantras of Preprocessing

### 1. Missing Data , Duplicated data, Replacement data , datatype setting correct
### 2. Encoding according to the categorical variables , ordinal or nominal
### 3. Outlier Treatment -> not required since we can see from X.describe()
### 4. Feature Scaling (Normalization or Standardization )
### 5. Imbalance Treatment (only applicable with Classification)


#### Normalization is best for algorithms sensitive to feature scales and data with bounded ranges,Standardization is best for algorithms assuming Gaussian-distributed data and data with outliers.,Always consider the distribution of the data and the requirements of the algorithm when choosing between normalization and standardization.

#### We need to do feature scaling when we have linear regression, SVM , etc. In sigmoid or if else things like decision trees it is not required

#### Multiple Model Comparison

```python

# logistic regression
logistic = LogisticRegression()
logistic=logistic.fit(X_train,y_train)
y_pred_lr = logistic.predict(X_test)
accuracy_lr = accuracy_score(y_test,y_pred_lr)

# Decision Tree
dec_tree = DecisionTreeClassifier()
dec_tree.fit(X_train,y_train)
y_pred_dec_tree=dec_tree.predict(X_test)
accuracy_dec_tree=accuracy_score(y_test,y_pred_dec_tree)

# Random Forest Classifier
random_fr = RandomForestClassifier()
random_fr=random_fr.fit(X_train,y_train)
y_pred_rf=random_fr.predict(X_test)
accuracy_rf=accuracy_score(y_test,y_pred_rf)

# XGBoost Classifier
xgboost = XGBClassifier()
xgboost=xgboost.fit(X_train,y_train)
y_pred_xgb=xgboost.predict(X_test)
accuracy_xgb=accuracy_score(y_test,y_pred_xgb)

# KNN
knc = KNeighborsClassifier(n_neighbors=3)
knc= knc.fit(X_train,y_train)
with threadpool_limits(limits=1, user_api="blas"):
    y_pred_knc = knc.predict(X_test)
accuracy_knc=accuracy_score(y_test,y_pred_knc)

# SVM
svm = SVC()
svm=svm.fit(X_train,y_train)
y_pred_svm=svm.predict(X_test)
accuracy_svm=accuracy_score(y_test,y_pred_svm)

# Naive Bayes
nbt = GaussianNB()
nbt=nbt.fit(X_train,y_train)
y_pred_nbt=nbt.predict(X_test)
accuracy_nbt=accuracy_score(y_test,y_pred_nbt)

# Voting Classifier - combining all the above model using voting classifier

model_eval = VotingClassifier(estimators=[('logistic',logistic),('dec_tree',dec_tree),('random_fr',random_fr),('xgboost',xgboost),('knc',knc),('svm',svm),('nbt',nbt)],voting='hard')
model_eval=model_eval.fit(X_train,y_train)
pred_evc=model_eval.predict(X_test)
accuracy_evc=accuracy_score(y_test,pred_evc)

list_metrics=[]
list_accuracy=[]

list_models=['Logistic_regress','Decision_tree','Random_Forest','XGB','KNN','SVM','NBT','Voting']
list_accuracy=[accuracy_lr,accuracy_dec_tree,accuracy_rf,accuracy_xgb,accuracy_knc,accuracy_svm,accuracy_nbt,accuracy_evc]

df_accuracy = pd.DataFrame({'Method Used':list_models,'Accuracy':list_accuracy})
print(df_accuracy)

```
